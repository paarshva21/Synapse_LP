# **Standardization vs Normalization**

When processing data in a given numeric dataset, there are many types of variables in this dataset that have vast ranges. For example, for a Company, the dataset can have a column age ranging from 21-65 years, which isn’t too vast, but the same may not apply to Salary which will range over larger values (say, from Rs. 10,000-Rs. 1 crore).  Machine Learning algorithms work by calculating distance between these numerical values of variables. So, undue partiality may be accorded to variables having large ranges (like Salary), due to its numerically larger values, thus distorting the analysis.

Hence, these values need to be scaled to some standard values in a common range so as to allow the ML algorithm to give equal weightage to all types of variables. There are many ways to do this, two of them are Standardization and Normalization. They are explained below:


*   Standardization: Also called Z-score normalization, this method scales the values in such a way that its mean and standard deviation become 0 and 1 respectively. So, these two values are changed to a normal distribution, so the shape of the distribution isn’t affected. This is done by the formula:

                   x_n = (x_n - x_mean) / x_std

    It is useful when the ranges of values are defined and the distribution is Gaussian (bell curve). It is not affected by outliers in the dataset, and the output isn’t bounded to a certain range. It is used in optimization algorithms like gradient descent and in algorithms that compute distance between features like K-Nearest-Neighbors (KNN).
 
*   	Normalization: Also termed Min-Max normalization, this approach rescales features such that all of them fit in a range between (and including) 0 and 1. It is given by the formula:

                      x_n = (x_n - x_min) / (x_max - x_min)

    This method transforms the given dataset to a common scale. It is utilized when we don’t know the type of distribution of dataset and the values are of different scales. However, it does not deal with outliers well as large values (if present in dataset) too get assigned a value between 0 and 1. Artificial Neural Networks and Logistic Regression are some of the ML algorithms that use this technique.
